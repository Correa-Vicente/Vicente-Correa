Se cambia el nombre de cada columna para una mejor comprención

```python
df2 = df.rename(columns={'Constituency': 'Distrito', 'Const. No.':'N°Distrito','Leading Candidate':'CandidatoLíder','Leading Party':'PartidoLíder','Trailing Candidate':'ÚltimoCandidato','Trailing Party':'ÚltimoPartido','Margin':'Margen','Status':'Estado'})
df2
```

```python
filtro_1 = df2['ÚltimoCandidato'].isna()
df2[filtro_1]
```

```python
filtro_2 = df2['Distrito']=='Surat'
df2[filtro_2]
```

Ambos datos nulos se encuentran en la sección de las elecciones para el distrito Surat

Tras investigar, se encontró que los datos se encuentran nulos debido a que todos los otros candidatos abandonaron las elecciones en el distrito debido a irregularidades en el proceso electoral y amaños en las firmas del candidato del congreso para el distrito, dejando a MUKESHKUMAR CHANDRAKAANT DALAL como único candidato y vencedor, por ende se sustituiran las celdas NaN por 'Sin Opositor'.
Fuente : https://www.livemint.com/elections/lok-sabha-election-result-2024-surat-constituency-already-in-bjps-pocket-heres-how-11717467784497.html

```python
df2['ÚltimoCandidato']=df2['ÚltimoCandidato'].fillna('Sin Oposición')
df2['ÚltimoPartido']=df2['ÚltimoPartido'].fillna('Sin Oposición')
filtro_4=df2['Distrito']=='Surat'
df2[filtro_4]
```

```python
df2.isna().sum()
```

Se toma como muestra los 10 primeros distritos para mostrar el Margen de votos de cada partido ganador

```python
df3 = df2.head(10)
```

Se convierten los datos de 'Margen' a numéricos ya que pandas los estaba tomando como cadenas en lugar de números y no calzaba la altura de las barras en el gráfico

```python
df3['Margen'] = pd.to_numeric(df3['Margen'], errors='coerce')
```

A continuación se muestra un gráfico que detalla el margen por el cual ganó cada candidato en las elecciones en India para los primeros 10 distritos

```python
plt.figure(figsize=(14, 8))

plt.bar(df3['CandidatoLíder'], df3['Margen'], color='blue', label='Líder')

plt.xlabel('Candidatos')
plt.ylabel('Margen')
plt.title('Margen de victoria de Candidatos Ganadores Primeros 10 distritos')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

A continuación se muestra un gráfico de la cantidad de victorias por partido en los primeros 20 distritos

```python
df4=df2.head(20)

victorias_por_partido = df4['PartidoLíder'].value_counts()

plt.bar(victorias_por_partido.index, victorias_por_partido.values, color='skyblue')

plt.xlabel('Partido')
plt.ylabel('Cantidad de Victorias')
plt.title('Cantidad de Victorias por Partido')

plt.xticks(rotation=45)

plt.subplots_adjust()

plt.show()
```
